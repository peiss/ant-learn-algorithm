## 哈希表

特点：查找速度是O(1)  
用于：空间换时间

常常用于缓存，然后快速查找，提升性能

【问题描述】无序数组中找出和为N的两个数，例如，nums = [1, 4, 3, 2, 6, 5]中找出和为target = 6的序列，答案：[(1, 5), (4, 2)]。

```python
def find(num_list, sum_value):
    # 第一遍历，O(N)，建立哈希表
    hash_dict = {}
    for val in num_list:
        if val not in hash_dict:
            hash_dict[val] = 0
        hash_dict[val] += 1
    print(hash_dict)
    
    # 第二次遍历，O(N)
    # 遇到每个元素val，判断sum_value-val在不在哈希表中
    for val in num_list:
        if hash_dict[val] == 0:
            continue
        hash_dict[val] -= 1
        if hash_dict.get(sum_value-val, 0) > 0:
            print(val, sum_value-val) 
            hash_dict[sum_value-val] -= 1
```

```python
num_list = [1, 4, 3, 2, 3, 4, 6, 5]
find(num_list, 6)
```
